#합성곱
합성곱과 밀집층의 차이
- 밀집층은 모든 특성 -> n개의 뉴런 -> 각 뉴런마다 다른 가중치를 곱해서 -> 뉴련마다  1개의 출력
- 합성곱은 필터에 맞는 갯수의 특성 -> 하나의 필터로 n개의 출력 -> 이 출력을 위치에 맞게 놓은 것 -> 특성 맵 -> 필터가 n개일때 n개의 특성맵 출력

- 뉴런 마다 가중치가 다른 밀집층과 다르게 합성곱은 하나의 필터에 동일한 가중치로 n개의 출력을 만듬

# 패딩 과 스트라이드
###1. 패딩
- **정의** : 입력 배열 주위에 가상의 원소로 채우는 것
- **하는 이유** : 모서리의 정보를 잃어버리지 않기 위해,특성 맵의 크기가 입력의 크기와 동일하게 유지하기 위해
1. 기본값 : valid 패딩 -> 패딩을 하지 않음
2. same 패딩 : 주위를 0으로 감싸는 패딩 (0으로 감싸기에 계산에 영향 x)


###2. 스트라이드
-**정의** : 필터가 이동하는 크기
- 기본값 : 1
- strides 매개변수 사용

즉 코드상으로 

keras.layers.Conv2D(필터 갯수, 커널 크기, 활성화 함수, pedding ='~', strides = n) 같은 형태
