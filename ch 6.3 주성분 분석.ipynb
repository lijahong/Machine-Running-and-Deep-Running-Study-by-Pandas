```python
import numpy as np
fruits = np.load('fruits_300.npy')
fruits_2d = fruits.reshape(-1,100*100)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 50)
pca.fit(fruits_2d)
```

```python
print(pca.components_.shape)
```

```python
import matplotlib.pyplot as plt
def draw_fruits(arr, ratio = 1):
  n = len(arr)
  rows = int(np.ceil(n/10))
  cols = n if rows <2 else 10
  fig, axs = plt.subplots(rows,cols, figsize = (cols *ratio, rows* ratio), squeeze = False)
  for i in range(rows):
    for j in range(cols):
      if i*10 +j <n:
        axs[i,j].imshow(arr[i*10 +j],cmap = 'gray_r')
      axs[i,j].axis('off')
  plt.show()
```

```python
draw_fruits(pca.components_.reshape(-1,100,100))
```

```python
print(fruits_2d.shape)
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
print(fruits_inverse.shape)
```

```python
fruits_a = fruits_inverse.reshape(-1,100,100)
for i in [0,100,200]:
  draw_fruits(fruits_a[i:i+100])
  print("\n")
```

```python
print(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
target = np.array([0]*100 + [1] * 100 + [2]*100)
```

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(lr, fruits_2d,target)
scores2 = cross_validate(lr,fruits_pca,target)
scores3 = cross_validate(lr,fruits_inverse,target)

```

```python
print(np.mean(scores['test_score']))
print(np.mean(scores2['test_score']))
print(np.mean(scores3['test_score']))
```

```python
pca = PCA(n_components=0.5)
pca.fit(fruits_2d)
print(pca.n_components_)
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

```python
scores4 = cross_validate(lr,fruits_pca,target)
print(np.mean(scores4['test_score']))
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters =3, random_state = 42)
km.fit(fruits_pca)
print(np.unique(km.labels_, return_counts = True))
```

```python
for label in range(0,3):
  draw_fruits(fruits[km.labels_ == label])
  print("\n")
```

```python
draw_fruits(fruits[km.labels_ == 0])
print("\n")
```

```python
for i in range(0,3):
  data = fruits_pca[km.labels_ == i]
  plt.scatter(data[:,0],data[:,1])
plt.legend(['apple','banana','pineapple'])
plt.show()
```
